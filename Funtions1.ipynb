```python
def OddEven():
    num = int(input('Enter a number : '))
    if(num % 2 != 0):
        print(num, 'is an oddd number')
    else:
        print(num, ' is an even number')
```

```python
#Your Gender:Male
#Your Age:20
#NOT ELIGIBLE
def Elegible():
    gen = input('Your Gender: ')
    age = int(input('Your age: '))
    if((gen =='Male' and age<21) or (gen =='Female' and age<18)):
        print('Not ELIGIBLE')
    else:
        print('ELIGIBLE')
```

```python
#Subject1= 98
#Subject2= 87
#Subject3= 95
#Subject4= 95
#Subject5= 93
#Total : 468
def percentage():
    s1 = int(input('Subject1 : '))
    s2 = int(input('Subject2 : '))
    s3 = int(input('Subject3 : '))
    s4 = int(input('Subject4 : '))
    s5 = int(input('Subject5 : '))
    tm = s1+s2+s3+s4+s5
    pcent = (tm/500)*100
    print('Total: ' , tm)
    print('Percentage: ',pcent)
```
