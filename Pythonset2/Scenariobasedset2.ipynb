```python
#cenario: You are developing a banking application that categorizes transactions based on the amount entered.
#Write logic to determine whether the amount is positive, negative, or zero.
```

```python
Acountbalance=1000
#1)Deposit
#2)Withdraw
#3)Balance enquiry
Type=input("What you would like to do?")
if Type =='1':
    da=int(input("Enter the amount to be deposited"))
    newbalance = Acountbalance + da
    print("Thanks for banking!Account balance: %d " %(newbalance))
    print("The transcation is possitive")
elif(Type=='2'):
    wda=int(input("Enter the amount to be withdrawn"))
    if wda<Acountbalance:
        newbalance = Acountbalance - wda
        print("Thanks for banking!Account balance: %d " %(newbalance))
        print("The transcation is negative")
    else:
        print("Insufficient balance")
        print("The transaction is Zero")
elif(Type=='3'):
    print("Your account balance is :",Acountbalance)
    print("The transaction is Zero")
else:
    print("Invalid option")
```

```python
#Scenario: In a secure login system, certain features are enabled only for users with prime-numbered user IDs.
#Write logic to check if a given number is prime.
```

```python
#Scenario: A scientist is working on permutations and needs to calculate the factorial of numbers frequently.
#Write logic to find the factorial of a given number using recursion.
#recursion -- function calling itself
```

```python
num = int(input("enter the number to find factorial"))
def factorial(x):
    inti=1
    for i in range(1,x+1):
        inti=inti*i
    return inti
print(factorial(num))
```

```python
#Scenario: A unique lottery system assigns ticket numbers where only Armstrong numbers win the jackpot.
#Write logic to check whether a given number is an Armstrong number.
```

```python
lotno= input("Enter your lottery number")
start=0
for i in lotno:
    start=start+(int(i))**3
if int(lotno) == start:
    print("The number is armstrong number")
else:
    print("The number is not armstrong number")
```

```python
#Scenario: A password manager needs to strengthen weak passwords by swapping the first and last characters of user-generated passwords.
#Write logic to perform this operation on a given string.
```

```python
pwd= input("Enter your password")
lst=[]
for i in pwd:
    lst.append(i)
firstchar=lst[0]
lastchar=lst[len(lst)-1]
newlist = lst.copy()
newlist[0] = lastchar
newlist[len(newlist)-1]=firstchar
print("Strengthened password is:" , "".join(newlist))
    
```

```python
#Scenario: A text-processing tool helps summarize articles by identifying the most significant words.
#Write logic to find the longest word in a sentence
```

```python
sentence = input("enter the sentence").split()
print(sentence)
lst=[]
for i in sentence:
    lst.append(len(i))
print(lst)
largest=lst[0]
for items in lst:
    if items > largest:
        largest=items
    ind = lst.index(largest)
print(sentence[ind])
```

```python
#Scenario: A plagiarism detection tool compares words from different documents and checks if they are anagrams (same characters but different order).
```

```python
reference = "comprehension is the reason for reading. If readers can read the words but do not understand or connect to what they are reading, they are not really reading. Good readers are both purposeful and active, and have the skills to absorb what they read, analyze it, make sense of it, and make it their own"
reference_list = reference.split()
print(reference_list)
sentence = input("enter the sentence").split()
print(sentence)
count=0
for it1 in sentence:
    if it1 in reference_list:
        count=count+1
print(count)
itisthere = False
for s in sentence:
    for r in reference_list:
        if sorted(s.lower()) == sorted(r.lower()):
            print("Anagram found:", s, r)
            itisthere = True
            break
    if itisthere:
        break

if not itisthere:
    print("Plagiarism not detected")
    
```
