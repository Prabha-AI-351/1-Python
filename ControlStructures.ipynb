```python
#if-else
age=int(input('enter the age'))
if(age<18):
    print('child')
else:
    print('adult')
```

```python
#if-elif
age=int(input('enter the age'))
if(age<18):
    print('child')
elif(age<35):
    print('adult')
elif(age<59):
    print('citizn')
else:
    print('senior citizen')
```
